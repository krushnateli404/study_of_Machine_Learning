```python
x = df.drop(['UDI','Product ID','Type','Air temperature [K]'],axis = 1)
x
```

```python
y = df['Air temperature [K]']
df2 = pd.DataFrame(y)
df2
```

```python
scalar = StandardScaler()
```

```python
arr = scalar.fit_transform(x)
arr
```

```python
df2 = pd.DataFrame(arr)
df2
```

```python
df2 = df2.rename(columns={0:'Process temperature [K]',1:'Rotational speed [rpm]',2:'Torque [Nm]',3:'Tool wear [min]',4:'Machine failure',5:'TWF',6:'HDF',7:'PWF',8:'OSF',9:'RNF'})
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif_df = pd.DataFrame()
```

```python
vif_df['VIF'] = [variance_inflation_factor(arr,i) for i in range(df2.shape[1])]
```

# Note:Machine failure column is a multi-collinear so we remove this columns cause vif value is greater than 10 and if we remove this column so our building modek complexity is decrease

```python
df2 =df2.drop(columns = 'Machine failure')
df2

```

```python
x_train,x_test,y_train,y_test = train_test_split(df2,y,test_size=0.25,random_state=200)
```

```python
x_train
```

```python
y_train
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.predict([[-0.947360,-0.958170 ,0.763720 ,0.943411 ,-0.06798,-0.10786,-0.097934,-0.099484 ,-0.04363]])
```

```python
# pickle.dump(lr,open('lr_model.sav','wb'))
# lr_model_save = pickle.load(open('lr_model.sav','rb'))

pickle.dump(lr,open('lr_model_1.sav','wb'))
lr_model_save = pickle.load(open('lr_model_1.sav','rb'))
```

```python
lr_model_save.score(x_test,y_test)
```
