```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/amberkakkar01/Prediction-of-Wine-Quality/master/winequality-red.csv')
data
```

```python
x = data.drop(columns='quality')
x
```

```python
y = data['quality']
y
```

```python
train,val_train,test,val_test = train_test_split(x,y,random_state=30,test_size=.50)
```

```python
x_train,x_test,y_train,y_test = train_test_split(train,test,test_size=.20,random_state=30)
```

```python
knn = KNeighborsClassifier()
svc = SVC()
```

```python
knn.fit(x_train,y_train)
svc.fit(x_train,y_train)
```

```python
knn.score(x_test,y_test)
```

```python
svc.score(x_test,y_test)
```

```python
prediction_knn = knn.predict(val_train)
prediction_svc = svc.predict(val_train)
```
