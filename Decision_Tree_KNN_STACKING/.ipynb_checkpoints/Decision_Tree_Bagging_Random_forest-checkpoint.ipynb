```python
df1 = df.head(800)
df1
```

```python
x1 = df1.drop(columns='quality')
```

```python
y1 = df1.quality
```

```python
x_train,x_test,y_train,y_test = train_test_split(x1,y1,test_size=.25,random_state=100)
```

```python
dt1_model = DecisionTreeClassifier()
dt1_model.fit(x_train,y_train)
```

```python
set(x1)
```

```python
plt.figure(figsize = (20,20))
tree.plot_tree(dt1_model,filled=True,class_names=[str(i) for i in set(x1)],feature_names=x1.columns)
plt.savefig('dt_model_1')
```

```python
[str(i) for i in set(x1)]
```

```python
df1
```

```python
# df1.profile_report()
```

```python
dt1_model.score(x1,y1)
```

```python
dt1_model.score(x_test,y_test)
```

```python
path = dt1_model.cost_complexity_pruning_path(x1,y1)
path
```

```python
ccp_alpha = path.ccp_alphas
ccp_alpha
```

```python
l1_model = []
for ccp in ccp_alpha:
    dt1 = DecisionTreeClassifier(ccp_alpha=ccp)
    dt1.fit(x_train,y_train)
    l1_model.append(dt1)
```

```python
l1_model
```

```python
train_score = [i.score(x1,y1) for i in l1_model]
train_score
```

```python
test_score = [i.score(x_test,y_test) for i in l1_model]
test_score
```

```python
fig,ax = plt.subplots()
ax.set_xlabel("alpha")
ax.set_ylabel("accuracy")
ax.set_title("Accuracy vs alpha for training and testing sets")
ax.plot(ccp_alpha ,train_score,marker = 'o',label = 'train',drawstyle = 'steps-post')
ax.plot(ccp_alpha , test_score,marker = 'o',label = 'test',drawstyle = 'steps-post')
ax.legend()
plt.show()
```
